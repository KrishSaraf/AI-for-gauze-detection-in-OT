```python
import cv2

# Replace 'your_video.mp4' with the path to your video file.
video_path = 'C:\\Users\\user\\yolov5-master\\frame.mp4'


cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Set the start time to 3 minutes (180 seconds)
start_time = 180  # 3 minutes
frame_rate = 720  # Frame extraction rate in seconds

# Seek to the start time
cap.set(cv2.CAP_PROP_POS_MSEC, start_time * 1000)
x=0
while True:
    ret, frame = cap.read()
    if x==frame_rate:
        if not ret:
            break

        # Save the frame to your computer
        frame_file_name = f"frame_{int(cap.get(cv2.CAP_PROP_POS_FRAMES))}.jpg"
        cv2.imwrite(frame_file_name, frame)
        print(f"Saved {frame_file_name}")

        cv2.imshow("Frame", frame)
    else:
        x+=1
    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import os

# Open the video file
video_file = 'C:\\Users\\user\\Desktop\\Video 2\\WIN_20231005_18_26_48_Pro.mp4'  # Replace with the path to your video file
cap = cv2.VideoCapture(video_file)

# Check if the video file was opened successfully
if not cap.isOpened():
    print("Error: Could not open video file")
    exit()

# Set the frame rate and time interval for frame extraction
fps = 30.02  # Frames per second of the video
time_interval = 5.0  # Seconds

# Calculate the frame interval based on the desired time interval
frame_interval = int(fps * time_interval)

frame_count = 0

# Create the "images" folder on your desktop if it doesn't exist
desktop_path = os.path.join(os.path.expanduser('~'), 'Desktop')
images_folder = os.path.join(desktop_path, 'images')
os.makedirs(images_folder, exist_ok=True)

while True:
    ret, frame = cap.read()

    if not ret:
        break  # End of video

    frame_count += 1

    # Extract a frame after every 5 seconds
    if frame_count % frame_interval == 0:
        # Save the frame to the "images" folder on your desktop
        frame_filename = os.path.join(images_folder, f'frame_{frame_count}.jpg')
        cv2.imwrite(frame_filename, frame)
        print(f'Saved frame {frame_count} as {frame_filename}')

# Release the video capture object
cap.release()
```
